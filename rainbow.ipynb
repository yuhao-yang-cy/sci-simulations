```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
index_profile = {'red':1.331, 'orange':1.332, 'yellow':1.333, 'green': 1.335, 'blue': 1.338, 'violet':1.342}
```

```python
def find_all(start, end, r_index):
    start *= math.pi / 180
    end *= math.pi / 180
    alpha = np.arange(start, end, 0.01)
    beta = np.arcsin( np.sin(alpha) / r_index )
    theta = (4*beta - 2*alpha) / math.pi * 180
    return theta
```

```python
def find_max(r_index):
    angles = find_all(0, 90, r_index)
    return max(angles)

find_max(1.33)
```

```python
fig, ax = plt.subplots()

a = 32
b = 80

alpha = np.arange(a*math.pi/180, b*math.pi/180, 0.01) / math.pi * 180

red_curve = ax.plot(alpha, find_all(a, b, index_profile['red']), 'red', label='red')
orange_curve = ax.plot(alpha, find_all(a, b, index_profile['orange']), 'orange', label='orange')
yellow_curve = ax.plot(alpha, find_all(a, b, index_profile['yellow']), 'yellow', label='yellow')
green_curve = ax.plot(alpha, find_all(a, b, index_profile['green']), 'green', label='green')
blue_curve = ax.plot(alpha, find_all(a, b, index_profile['blue']), 'blue', label='blue')
violet_curve = ax.plot(alpha, find_all(a, b, index_profile['violet']), 'violet', label='violet')

ax.legend()
plt.ylabel('Angle of Incidence')
plt.xlabel('Deflected Angle')
plt.show()
```
